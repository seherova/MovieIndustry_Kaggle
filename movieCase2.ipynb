# Hipotez 1: Bütçesi fazla olanların Gişe Hasılatı da fazladır.

H0: Budget'ı fazla olanların Gross'u da fazla değildir.

H1: Budget'ı fazla olanların Gross'u da fazladır..

İlk önce normal dağılıma uyuyor mu diye test ediyoruz, eğer normal dağılıma uyuyorsa İndependent T Test'i yapıyoruz eğer uymuyorsa İndependent T Test'inin non-parametrik karşılığı olan Mann Whitney U Testini yapabiliriz ya da korelasyon kullanabiliriz.

normal dağılıma uyuyor mu uymuyor mu diye bakmak için shapiro kullanıyoruz (scipy-stats)

```python
from scipy import stats

statistic, p_value= stats.shapiro(df['gross'])

alpha= 0.05

if p_value< alpha:
    print("Data don't fit normal distribution.")
else:
    print("Data fit normal distribution")
```

**Non-Parametrik Test: Mann Whitney U**

```python
from scipy.stats import mannwhitneyu


statistic, p_value = mannwhitneyu(df["budget"], df["gross"])

alpha = 0.05

print("Statistic: ", statistic)
print("p value: ", p_value)

if p_value < alpha:
    print("We reject the H0 hypothesis. Those with a large budget have a high gross.")
else:
    print("We accept the H0 hypothesis. The gross of those with a large budget is not high")
```

# Hiptez 2: Süresi fazla olan filmlerin scor’u düşüktür. 

H0: Runtime'ı yüksek olan filmlerin scor'u düşük değildir.

H1: Runtime'ı yüksek olan filmlerin scor'u düşüktür.

```python
#df['runtime'] = df['runtime'].astype(float)

# Shapiro-Wilk testini uygulayalım
statistic, p_value = stats.shapiro(df['runtime'])

alpha = 0.05

print("Statistic:", statistic)
print("p value:", p_value)

if p_value < alpha:
    print("Data do not fit normal distribution.")
else:
    print("Data fit normal distribution.")
```

**Correlation solution for non-parametric: Spearmanr**

```python
from scipy.stats import spearmanr

# Spearman's Rho testini uygulayın
correlation, p_value = spearmanr(df['runtime'], df['score'])

alpha = 0.05

print("Spearman's Rho correlation coefficient: ", correlation)
print("p value: ", p_value)

if p_value >= alpha:
    print("H0 hypothesis is true. Films with a high runtime do not have a low score, the duration of the movie does not affect the score in this way.")
else:
    print("We reject the H0 hypothesis. Movies with a high runtime have a low score. Prolonging the duration of the film negatively affects its score.")
```

```python
correlation_runtime_and_score= df.corr(method='spearman')
sns.heatmap(correlation_runtime_and_score, annot= True)
```

**Correlation with Kendall**

```python
correlation_runtime_and_score= df.corr(method='kendall')
sns.heatmap(correlation_runtime_and_score, annot= True)
```

```python
#parametrik olmayan ilişki inceliyoruz bu sebeple Kendall Tau kullanacağım.
from scipy import stats

correlation_coef, p_value= stats.kendalltau(df['runtime'], df['score'])

alpha= 0.05

if p_value< alpha:
    print("We reject H0, movies with high Runtime have low score.")
else:
    print("We accept H0, movies with high Runtime do not score low.")
```

# Hipotez 3: 2000'den sonraki filmlerin votes'ı öncesine göre daha fazladır.

H0: Year'ı 2000'den büyük olan filmlerin votes'ı daha fazla değildir.

H1: Year'ı 2000'den büyük olan filmlerin votes'ı daha fazladır.

```python
statistic, p_value = stats.shapiro(df['votes'])

alpha = 0.05

print("Statistic:", statistic)
print("p value:", p_value)

if p_value < alpha:
    print("Data do not fit normal distribution.")
else:
    print("Data fit normal distribution.")
```

**Non-parametric Test: Mann Whitney U Test**

```python

moviesAfter_2000 = df[df['year'] >= 2000]
moviesBefore_2000 = df[df['year'] < 2000]

statistic, p_value = mannwhitneyu(moviesAfter_2000["votes"], moviesBefore_2000["votes"])

alpha = 0.05

print("Statistic: ", statistic)
print("p value: ", p_value)

if p_value >= alpha:
    print("The H0 hypothesis accepted. Movies after 2000 have no more votes.")
else:
    print("The H0 hypothesis reject. Films after 2000 have more votes..")
```

```python
import matplotlib.pyplot as plt

average_votes = df.groupby('year')['votes'].mean()

plt.plot(average_votes.index, average_votes.values)
plt.xlabel('Year')
plt.ylabel('Average Votes')
plt.title('Average Votes Values by Years')
plt.grid(True)
plt.show()
```

**Non-parametric Test: Mann Whitney U Test**

# Hipotez 7: Yüksek score'lar genel olarak US'tedir.

H0: Yüksek score'lar genel olarak US'de değildir.

H1: Yüksek score'lar genel olarak US'tedir.

**Non-Parametric Test:**

```python
us_Movies = df[df['country'] == 'United States']
other_Country = df[df['country'] != 'United States']

statistic, p_value = mannwhitneyu(us_Movies["score"], other_Country["score"])

alpha = 0.05

print("Statistic: ", statistic)
print("p value: ", p_value)

if p_value < alpha:
    print("H0 hipotezini reddediyoruz. Yüksek score'lar genel olarak US'tedir.")
else:
    print("H0 hipotezini kabul ediyoruz. Yüksek score'lar genel olarak US'te değildir.")
```

# Hipotez 8: Aksiyon filmleri dramadan daha fazla gişe hasılatın sahiptir.

H0: Aksiyon filmleri dramadan daha fazla grossa(gişe hasılatı) sahip değildir.

H1: Aksiyon filmleri dramadan daha fazla grossa(gişe hasılatı) sahiptir.

```python
import seaborn as sns

action_Movies= df[df['genre']== 'Action']
drama_Movies= df[df['genre']== 'Drama']

sns.boxplot(x= 'genre', y= 'gross', data= pd.concat([action_Movies, drama_Movies]))
plt.xlabel('Genre')
plt.ylabel('Gross')
plt.title('The Gross Distribution of Action and Drama Films')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Yalnızca Action ve Drama filmlerini filtreleyelim
action_Movies = df[df['genre'] == 'Action']
drama_Movies = df[df['genre'] == 'Drama']

# Action ve Drama filmlerinin gişe hasılatlarının ortalamasını hesaplayalım
action_mean_gross = action_Movies['gross'].mean()
drama_mean_gross = drama_Movies['gross'].mean()

# Çubuk grafiği ile Action ve Drama filmlerinin gişe hasılatlarını karşılaştıralım
plt.bar(['Action', 'Drama'], [action_mean_gross, drama_mean_gross], color=['blue', 'orange'])
plt.xlabel('Genre')
plt.ylabel('Average Gross')
plt.title('The Gross Distribution of Action and Drama Films')
plt.show()
```

**Non-parametric Test: Mann Whitney U**

```python
from scipy.stats import mannwhitneyu

action_Movies = df[df['genre'] == 'Action']
drama_Movies = df[df['genre'] == 'Drama']

statistic, p_value = mannwhitneyu(action_Movies["gross"], drama_Movies["gross"])

alpha = 0.05

print("Statistic: ", statistic)
print("p value: ", p_value)

if p_value < alpha:
    print("We reject the H0 hypothesis. Action movies have more gross than drama.")
else:
    print("We accept the H0 hypothesis. Action movies don't have more gross than drama.")
```

# Hipotez 9:Bütçeleri Yüksek Olan Filmlerin Vote'ları da Yüksektir

H0: Filmlerin bütçeleri ile Vote'ları arasında ilişki yoktur.

H1: Filmlerin bütçeleri ile Vote'ları arasında pozitif bir ilişki vardır.

```python
statistic, p_value = stats.shapiro(df['budget'])

alpha = 0.05

print("Statistic:", statistic)
print("p value:", p_value)

if p_value < alpha:
    print("Data do not fit normal distribution.")
else:
    print("Data fit normal distribution.")
```

**Correlation with Kendall Tau**

```python
correlation_coef, p_value= stats.kendalltau(df['budget'], df['votes'])

print("correlation_coef: ", correlation_coef)
print("p_value: ", p_value)

alpha= 0.05

if p_value< alpha:
    print("We reject the H0. There is a positive relationship between the budgets of the films and their Votes.")
else:
    print("We accept the H0. There is no relationship between the budgets of the movies and their Votes.")
```
